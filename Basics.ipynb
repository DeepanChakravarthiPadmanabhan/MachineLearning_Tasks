## Task 1. (Introduction)
Suppose that we have three coloured boxes r (red), b (blue), and g (green).

- Box r contains 3 apples, 4 oranges, and 3 limes, 
- Box b contains 1 apple, 1 orange, and 0 limes, and 
- Box g contains 3 apples, 3 oranges, and 4 limes. 

If a box is chosen at random with probabilities p(r) = 0.2, p(b) = 0.2, p(g) = 0.6, and a piece of
fruit is removed from the box (with equal probability of selecting any of the items in
the box), then what is the probability of selecting an apple? 

If we observe that the selected fruit is in fact an orange, what is the probability that it came from the green
box?

## Given data:

* Probability(choosing red box)= P(r)=0.2

* Probability(choosing blue box)= P(b)=0.2

* Probability(choosing green box)= P(g)=0.6

* P(a), P(o) and P(l) be the prior probability of finding apples, oranges and limes in the boxes.

## Mathematical formulations:

* Coonditional probability involving two events A and B, where Probability of A given by is given as:
\begin{align}
P(A|B)= \frac{P(A\cap B)}{P(B)}
\end{align}

where A and B are two events and P(B)$\neq$0.

* Bayes rule is given by:
\begin{align}
P(A|B)=\frac{P(B|A)\times P(A)}{P(B)}
\end{align}

where A and B are two events and P(B)$\neq$0.

In this equation,

P(A|B) is a conditional probability and it is the likelihood of event A occuring given that B is true.

P(B|A) is a conditional probability and it is the likelihood of event B occuring given that A is true.

P(A) and P(B) are the probabilities of observing A and B independently of each other ie., Prior probabilities.


## Solution:

$P(a) = P(a \cap r) + P(a \cap b) + P(a \cap g)$

From the product rule stated below,

\begin{align}
P(A \cap B)= P(A|B) \times P(B)
\end{align}

where A and B are two events and P(B) $\neq$0 and P(A|B) is a conditional probability and it is the likelihood of event A occuring given that B is true.

P(A $\cap$ B) is the probability of event A and B occuring together. 

Substituting product rule for P(a),

$ P(a) = (P(a|r) \times P(r)) + (P(a|b) \times P(b)) +(P(a|g) \times P(g)) $

$P(a) = 0.2 \times \frac{3}{10} + 0.2 \times \frac{1}{2} + 0.6 \times \frac{3}{10}$

$P(a) = 0.34$

#### The probability of selecting an apple is 0.34.

$P(g|o) = \frac{P(g \cap o)}{P(o)}$

$P(g \cap o)= P(g) \times P(o|g)$

$P(g \cap o)= 0.6 \times \frac{3}{10}$

$P(g \cap o)= 0.18$

$P(o) = P(o \cap g) + P(o \cap b) + P(o \cap r)$

$P(o) = P(r) \times P(o|r) + P(b) \times P(o|b) + P(g) \times P(o|g)$

$P(o) = 0.2 \times \frac{4}{10} + 0.2 \times \frac{1}{2} + 0.6 \times \frac{3}{10} $

$P(o) = 0.36$

Substituting on P(g|o),

$P(g|o)= \frac{0.18}{0.36}$

$P(g|o) = 0.5$

#### The probability of an orange coming from the green bag is 0.5.


## Task 2. (Introduction)
Consider a Gaussian distribution $p_x(x)$ over x with mean Î¼ = 6 and
standard deviation $\sigma$ = 1. Plot this distribution.  
Next, draw a sample of N = 50000 points from this distribution and plot a histogram of their
values, which as expected agrees with the distribution $p_x(x)$.  
Now consider a non-linear change of variables from x to y given by  
$x=g(y)=ln(y)-ln(1-y)+5$  
The inverse of this function is given by  
$y=g^{-1}(x)=\frac{1}{1+exp(-x+5)}$  
which is a logistic sigmoid function. Plot this function in the same Figure.  
Plot the curve $p_x(g(y))$ that is a transformation of $p_x(x)$ as a function of x $p_x(g(y))$ 

```python
def inverse_sigmoid(y):
    return np.log(y)-np.log(1-y)+5

def sigmoid_plot(x):
    return 1/(1+np.exp(-x+5))

# Parameter initialization
f = plt.figure(figsize=(12,9))
plt.subplots_adjust(left=None,bottom=0.1,top=0.9,wspace=0.2,hspace=0.5)
number_of_samples=500
mean=6
std_dev=1

# Plot for the histogram
plt.subplot(121)
x=np.random.normal(mean,std_dev,50000)
counts,bins,ignored=plt.hist(x,50,density=True,alpha=0.8,label='Histogram')
# The histogram obtained by 50000 samples from the normal distribution specified
plt.plot(bins,np.exp(-(bins - mean)**2 / 2*std_dev**2) / np.sqrt(2*np.pi*std_dev**2),color='red',label='Normal distribution')
# The normal distribution specified
plt.title('Normal distribution and histogram')
plt.xlabel('Sample values(Bins)')
plt.ylabel('Probability')
plt.legend(loc='upper left')


plt.subplot(122)
x=np.linspace(0,10,number_of_samples)
normal = np.exp(-(x - mean)**2 / 2*std_dev**2) / np.sqrt(2*np.pi*std_dev**2)

# Normal distribution plot
plt.plot(x,normal,'orange',label='PDF')
plt.xlabel('Sample values')
plt.ylabel('Probability')
plt.title('Distribution-Function mapping')

# Sigmoid plot of the samples created
sigmoid_transformed=sigmoid_plot(x)
plt.plot(x,sigmoid_transformed,'red',label='Sigmoid')

# Inverse sigmoid and sigmoid mapping 
inverse=inverse_sigmoid(sigmoid_transformed)
plt.plot(normal*inverse,sigmoid_transformed,'blue',label='Mapping')
plt.legend()
plt.show()
```
